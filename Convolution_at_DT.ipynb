```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
s1 = np.ones(4, dtype=int)
s2 = np.ones(6, dtype=int)
```

```python
print("s1 : ", s1)
print("s2 : ", s2)
```

```python
x = np.array(range(0, s1.shape[0]))
plt.scatter(x, s1)
plt.grid(True)
plt.axis([-1, s1.shape[0]+1, np.min(s1)-1, np.max(s1)+1])
```

```python
x = np.array(range(0, s2.shape[0]))
plt.scatter(x, s2)
plt.grid(True)
plt.axis([-1, s2.shape[0]+1, np.min(s2)-1, np.max(s2)+1])
```

```python
y = np.zeros(s1.shape[0]+s2.shape[0]-1)
for i in range(0, s1.shape[0]+s2.shape[0]-1):
  if(i<s1.shape[0]):
    for k in range(0, i+1):
      y[i] = y[i] + s1[k]*s2[i-k]
  elif(i<s2.shape[0]):
    for k in range(0, s1.shape[0]):
      y[i] = y[i] + s1[k]*s2[i-k]
  else:
    for k in range(0, s1.shape[0]+s2.shape[0]-(i+1)):
      y[i] = y[i] + s1[i-s2.shape[0]+k+1]*s2[s2.shape[0]-(k+1)]
```

```python
y
```

```python
x2 = np.array(range(0, y.shape[0]))
plt.scatter(x2, y)
plt.grid(True)
plt.axis([-1, y.shape[0]+1, np.min(y)-1, np.max(y)+1])
```

```python
N = 4
M = 5
```

```python
random_array1 = np.zeros(N)
random_array2 = np.zeros(M)
```

```python
for i in range(0, N):
  random_array1[i] = random.randint(-5, 5)
for i in range(0, M):
  random_array2[i] = random.randint(-5, 5)

print(random_array1)
print(random_array2)
```

```python
r1 = np.array(range(0, random_array1.shape[0]))
plt.scatter(r1, random_array1)
plt.grid(True)
plt.axis([-1, random_array1.shape[0]+1, np.min(random_array1)-1, np.max(random_array1)+1])
```

```python
r2 = np.array(range(0, random_array2.shape[0]))
plt.scatter(r2, random_array2)
plt.grid(True)
plt.axis([-1, random_array2.shape[0]+1, np.min(random_array2)-1, np.max(random_array2)+1])
```

```python
y2 = np.zeros(random_array1.shape[0]+random_array2.shape[0]-1)
for i in range(0, random_array1.shape[0]+random_array2.shape[0]-1):
  if(i<random_array1.shape[0]):
    for k in range(0, i+1):
      y2[i] = y2[i] + random_array1[k]*random_array2[i-k]
  elif(i<random_array2.shape[0]):
    for k in range(0, random_array1.shape[0]):
      y2[i] = y2[i] + random_array1[k]*random_array2[i-k]
  else:
    for k in range(0, random_array1.shape[0]+random_array2.shape[0]-(i+1)):
      y2[i] = y2[i] + random_array1[i-random_array2.shape[0]+k+1]*random_array2[random_array2.shape[0]-(k+1)]
```

```python
y2
```

```python
r3 = np.array(range(0, y2.shape[0]))
plt.scatter(r3, y2)
plt.grid(True)
plt.axis([-1, y2.shape[0]+1, np.min(y2)-1, np.max(y2)+1])
```

```python
y4 = np.convolve(s1, s2)
y4
```

```python
x4 = np.array(range(0, y4.shape[0]))
plt.scatter(x4, y4)
plt.grid(True)
plt.axis([-1, y4.shape[0]+1, np.min(y4)-1, np.max(y4)+1])
```
