```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
b=0
x=np.arange(-1,1,0.0001)

for i in range(1,3):
  b=b+np.sin((4*i-2)*np.pi*x)/(2*i-1)

plt.plot(x,b)
plt.title('n=3')
plt.show()
```

```python
u=0
for i in range(1,10):
  u=u+np.sin((4*i-2)*np.pi*x)/(2*i-1)

plt.plot(x,u)
plt.title('n=10')
plt.show()
```

```python
q=0
for i in range(1,100):
  q=q+np.sin((4*i-2)*np.pi*x)/(2*i-1)

plt.plot(x,q)
plt.title('n=100')
plt.show()
```
